```python
import pandas as pd
import numpy as np

data = pd.read_csv('olympics.csv')
data.head()
```

```python
#Renaming column headers and dropping 1st row of data

data.columns = ['Country', 'Summer', '1', '2', '3', 'Total', 'Winter', '1', '2', '3', 'Total', 'Games', '1', '2', '3', 'Combined Total']
data = data.drop(0)
data.head()
```

```python
#dropping empty data from dataset using the dropna() function
data.dropna(inplace = True)
data.dtypes
```

```python
#converting string types to integers using .astype() to be able to do arithmetic if needed
data['Summer'] = data['Summer'].astype(int)
data['Winter'] = data['Winter'].astype(int)
data['Games'] = data['Games'].astype(int)
data['1'] = data['1'].astype(int)
data['2'] = data['2'].astype(int)
data['3'] = data['3'].astype(int)
data['Total'] = data['Total'].astype(int)
data['Combined Total'] = data['Combined Total'].astype(int)

data.dtypes
```

```python
#removed any duplicate countries in the dataset by using drop_duplicates()
data.drop_duplicates(inplace=True)
len(data)
```

```python
#finding the highest number of combined medal counts
data['Combined Total'].max()
print(data[data['Combined Total']==17579].index.values)
print(data.loc[[147]])

#removing total by drop
data = data.drop(147)
```

```python
data.sort_values('Combined Total', ascending=False, inplace=True)
data.index
data.head()
```
