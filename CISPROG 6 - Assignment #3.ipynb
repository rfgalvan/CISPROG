```python
import pandas as pd
import numpy as np

data = pd.read_csv("imdb.csv")
data.head(10)
```

```python
data.sort_values('Rating', ascending=False, inplace=True)
data.head(1)
```

```python
data.sort_values('Year', ascending=True, inplace=True)
data.head(1)
```

```python
data.sort_values('Rating', ascending=False, inplace=True)
data.head(20)
```

```python
print("The average rating of movies is ", np.average(data['Rating']))
```

```python
ninety_forties = data[data['Year'].between(1940,1949)]
print("The average rating of movies from the 1940's is ", np.average(ninety_forties['Rating']))
```

```python
ninety_nine = data[data['Year']==1999]
ninety_nine.head()
```

```python
really_highly_rated = data[data['Rating'] > 8.5]
really_highly_rated.head()
```

```python
avg_20th_century = data[data['Year'].between(1901,2000)]
print("The average rating of movies in the 20th Century is ", np.average(avg_20th_century['Rating']))

avg_21st_century = data[data['Year'].between(2001,2100)]
print("The average rating of movies in the 20th Century is ", np.average(avg_21st_century['Rating']))
```

```python

print("The porportion of movies made in the 20th Century is ", len(avg_20th_century)/len(data))

print("The porportion of movies made in the 21st Century is ", len(avg_21st_century)/len(data))
```
